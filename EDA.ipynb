### Extracting meaningful data from recent and overall review columns

```python

REGEX_PATTERN = re.compile('(.+),\((,|\d)+\),-\s(\d+)')
```

```python
match = REGEX_PATTERN.match("Very Positive,(554),- 89% of the 554 user...")
```

```python
match.group(0)
```

```python
match.group(1)
```

```python
match.group(2)
```

```python
match.group(3)
```

This is a test example that shows the groups that are of interest are `group 1`, and `group 3`

```python
def rx_match(row, group_number):
    try:
        result = REGEX_PATTERN.match(row).group(group_number)
        return result
    except:
        return ""

# x[x["recent_reviews"].isna()] = ""

# print(x)
```

```python



x["recent_review_class"] = x["recent_reviews"].apply(lambda row: rx_match(row, 1))
```

```python
x.head()
```

If you scroll to the right of the above result, you will notice that now we have an additional column named `recent_review_class`!

```python
# I changed the group row in the rx_match method parameter from 1 to 3 ------>
x["recent_review_score"] = x["recent_reviews"].apply(lambda row: rx_match(row, 3))

x.head()
```

Now I will go through the `genre` and `popular_tags` columns in order to find distinct values, and add a column for each of them so that the games of matching genre and tags can be recognized as similar.

This will have only 0s and 1s as values so I will separate it into a dataframe that will not necessarily be readable, but it will be used for finding similar items.

```python
def find_distinct_values(column):
    distinct_values = []
    for _, row in x.iterrows():
        if isinstance(row[column], float):
            continue
        values = row[column].split(",")
        for item in values:
            if item.lower() not in distinct_values:
                distinct_values.append(item.lower())
    return distinct_values
```

```python
find_distinct_values("genre")
```

```python
# numberic_data_df will have each popular tag and genre as columns, and values will be 0s or 1s depending on whether the given 
# row (game) contains that genre or tag.
numeric_data_df = pd.DataFrame()

def add_column(reference_column, new_column, function):
    if new_column in numeric_data_df.columns:
        print("Column " + new_column + " already exists...")
        return
    numeric_data_df[new_column] = x[reference_column].apply(lambda row: function(row, new_column))
    
```

```python
def contains(string, substring):
    if isinstance(string, str) and substring.lower() in string.lower():
        return 1
    return 0
```

```python
unwrapped_columns = ["genre", "popular_tags"] # Add more columns to do this for, such as popular_tags, publisher?

for column in unwrapped_columns:
    for distinct_value in find_distinct_values(column):
        add_column(column, distinct_value, contains)
```
